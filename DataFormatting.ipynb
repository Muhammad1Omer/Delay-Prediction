```python

from docx import Document
import json
import pandas as pd
import numpy as np
```

```python

all_flights = []
for i in range(1, 73):
    
    if i == 38:
        continue
    
    print(i)
    
    doc = Document(f"ML-Proj-Dataset\\Train\\{i}.docx")
    
    # Extract text from the document
    text = ""
    for paragraph in doc.paragraphs:
        text += paragraph.text

    # Parse the extracted text as JSON
    json_data = json.loads(text)
    all_flights.extend(json_data)
        
```

```python

def flatten_json(data, parent_key='', sep='_'):
    items = []
    for k, v in data.items():
        new_key = f"{parent_key}{sep}{k}" if parent_key else k
        if isinstance(v, dict):
            items.extend(flatten_json(v, new_key, sep=sep).items())
        else:
            items.append((new_key, v))
    return dict(items)
```

```python

flattened_data = [flatten_json(entry) for entry in all_flights]

df = pd.DataFrame(flattened_data)
print( len(df) )

df.head()
```

```python

output_file = "TrainFlight.csv"
df.to_csv(output_file, index=False)

print(f"Data successfully saved to {output_file}")
```

# Weather Data

```python

all_dataframes = []

year = 23

for i in range(1, 14):
    df = pd.read_excel(f"ML-Proj-Dataset\\Weather\\{i}.xlsx")
    
    df_time = df['Time'].str.split(' ', expand=True).iloc[0].dropna().reset_index(drop=True)

    if i == 7:
        year += 1

    for i in range(1, len(df_time)):
        df_time[i] = f'{df_time[i]}-{df_time[0]}-{year}'

    # Initialize the data dictionary
    columns_to_split = df.columns[1:]
    data = {'Time': df_time[1:] }

    # Process the mini columns
    for col in columns_to_split:
        split_data = df[col].iloc[0].split(' ')
        
        if split_data[0] in ['Max', 'Total']:
            if split_data[0] == 'Max':  # Max Avg Min
                data[f'{col}_Avg'] = split_data[1::3]
                data[f'{col}_Min'] = split_data[2::3]
                data[f'{col}_Max'] = split_data[3::3]
            elif split_data[0] == 'Total':  # Total
                data[f'{col}_Total'] = split_data[1:]
        else:
            raise ValueError(f"Unexpected header format in column: {col}")
        
    for col in data:
        data[col] = [value for value in data[col] if value not in ['Avg', 'Min']]

    result_df = pd.DataFrame(data)
    all_dataframes.append(result_df)

final_df = pd.concat(all_dataframes, ignore_index=True)

final_df.head()
```
